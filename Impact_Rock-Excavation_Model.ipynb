## Rock Breakdown and Regolith Formation

The solid surfaces of planetary bodies are typically made up of both loose regolith and blocky/consolidated materials. Regolith forms through the fragmentation of larger rocks by impact bombardment. Rocks can also be buried by unconsolidated impact ejecta materials. Production of new rocks is due to excavation from buried rocky layers, solidification of impact melt, and extrusive volcanism. Sintering of granular materials may also occur on icy bodies.

Here, we develop a model of rock breakdown and regolith formation, using a simple analytical approach. The rate of change of surface rock abundance can be written

\begin{equation}
    \frac{dR}{dt} = (\mathrm{Sources}) - (\mathrm{Losses})
\end{equation}

where $R$ is the fractional surface area occupied by rocks.

To begin with, we assume that the primary source of rocks is impact excavation. We also assume that the two dominant components of the loss term are: 1) fragmentation by meteorite impacts and 2) burial by impact ejecta. Below, we develop models for these terms in the rock abundance equation. First, we derive the dominant source term (impact excavation) and the first loss term (burial by impact ejecta) together (Section 1). Then, we develop a model for the rock breakdown term (Section 2).

### 1. Impacts: rock excavation and burial by ejecta

For simplicity, we start with a two-layer model, where impacts produce blocky ejecta when the crater penetrates to a rocky layer (Fig. 1). Ejecta from the volume above the rocky layer can bury rocks on the surface. This scenario may be applicable in the lunar maria, where a regolith layer has developed on top of a solid "bedrock" layer of solidified lava.

In this case, the emplaced volume of rocky ejecta, $V_\mathrm{R}$ depends on the regolith thickness, $z_\mathrm{r}$, and the crater depth, $d$. We write the proportion of rocky material as a function of these two parameters:

\begin{equation}
    f(z_\mathrm{r},d) = \frac{V_\mathrm{R}}{V_\mathrm{c}}
\end{equation}
where $V_\mathrm{c}$ is the total volume of the crater. We will find an exact formula for $f(z_\mathrm{r},d)$ below.

We assume that the ratio of surface area to volume of the rocks in the ejecta is equal to that of the whole ejecta blanket, i.e.,

\begin{equation}
    \frac{\mathrm{Area~of~Rocks}}{\mathrm{Volume~of~Rocks}} = \frac{\mathrm{Area~of~Ejecta}}{\mathrm{Volume~of~Ejecta}}
\end{equation}

In this case, the rock abundance in the ejecta blanket is

\begin{align}
    R_\mathrm{e} &= \frac{\mathrm{Area~of~rocks}}{\mathrm{Area~of~ejecta}} = \frac{\mathrm{Volume~of~rocks}}{\mathrm{Volume~of~ejecta}} \\
    &= f(z_\mathrm{r},d)
\end{align}

since the total volume of ejecta is equal the crater volume, $V_\mathrm{c}$.

#### 1.1 Rock production function

The area of the ejecta blanket is $A_\mathrm{e} = \pi r_\mathrm{max}^2$, with $r_\mathrm{max} = r_\mathrm{max}(D)$. We assume that the ejecta blanket size $r_\mathrm{max}$ depends only on the crater depth $d$, and that the crater depth depends only on the crater diameter, $D$ (e.g., $d/D = \mathrm{constant}$). Therefore $A_\mathrm{e} = A_\mathrm{e}(D)$ and $f = f(z_\mathrm{r},D)$. The surface area occupied by rocks is $A_\mathrm{rock} = AR$. Starting from an initial rock abundance $R$, the change in the rock area over the ejecta blanket is then

\begin{align}
    \Delta A_\mathrm{rock}(z_\mathrm{r},D,R) &= R_\mathrm{e}A_\mathrm{e} - R A_\mathrm{e} \\
    &= A_\mathrm{e}\left( R_\mathrm{e} - R \right) \\
    &= A_\mathrm{e}\left[ f(z_\mathrm{r},D) - R \right]
\end{align}

The area of crater ejecta area per unit time, per unit area, is determined by the production function for craters with diameter $D$:

\begin{align}
    N(>D) = \frac{\mathrm{Number~of~craters~larger~than}~D}{\mathrm{km^{2}~yr}}
\end{align}
or

\begin{align}
    n(D) &= \frac{dN}{dD} \\
    &\rightarrow \frac{\mathrm{Number~of~craters~from}~D~\mathrm{to~}(D+dD)}{\mathrm{km^{2}~yr}} = n(D)dD = \left(\frac{d}{dD}N(>D)\right)dD
\end{align}

We can write the rock production rate in terms of this production function:

\begin{align}
    \frac{\mathrm{Rock~area~produced~by~craters~from}~D~\mathrm{to~}(D+dD)}{\mathrm{km^{2}~yr}} = \Delta A_\mathrm{rock}(D) n(D)dD
\end{align}

and integrate to find the total rock area per unit area (i.e. rock abundance) per unit time:

\begin{align}
    \frac{dR}{dt} = \frac{\mathrm{Change~in~rock~area}}{\mathrm{km^{2}~yr}} = \boxed{\int_{D_\mathrm{min}}^{D_\mathrm{max}} \Delta A_\mathrm{rock}(z_\mathrm{r},D,R) n(D)dD}
\end{align}

#### 1.2  Volume of excavated rocks

Referring to Figure 1, we determine the volume of rocks, $V_\mathrm{R}$ excavated from a crater with depth $d$ and diameter $D$. We write the depth/diameter ratio $\beta = d/D$, and the total volume of a bowl-shaped crater is

\begin{align}
    V_\mathrm{c} &= \frac{\pi}{6}\beta D^3 \left( \frac{3}{4} + \beta^2 \right) \\
    &= \frac{1}{3}\pi d^2 \left(3r - d \right)
\end{align}

and the volume of the part of the crater below the regolith layer with depth $z_\mathrm{r}$ is

\begin{align}
    V_\mathrm{R} = \frac{\pi (d-z_\mathrm{r})^2}{3}(3a+z_\mathrm{r}-d)
\end{align}

where 

\begin{align}
    a = \frac{D^2 + 2d^2}{4d}
\end{align}

The fractional volume of rock is

\begin{align}
    f &= \frac{V_\mathrm{R}}{V_\mathrm{c}} \\
    &= \frac{(d-z_\mathrm{r})^2(3a+z_\mathrm{r}-d)}{d^2(3a-d)}
\end{align}

For a constant crater depth/diameter ratio $\beta$, we can substitute $d = \beta D$. Alternatively, $\beta$ can be a function of $D$. In either case, we can write $f$ in terms of $D$ and $z_\mathrm{r}$ only.

#### 1.3  Extent and area of ejecta

Ejecta thickness decreases rapidly with distance from the crater rim. This thickness can be modeled using a power-law,

\begin{align}
    h(r) = C\left(\frac{r}{D}\right)^{-B}
\end{align}

typically with $B = 3$. The coefficient $C$ can be found by conservation of mass, assuming no density change in the ejecta:

\begin{align}
    V_\mathrm{c} = \int dV &= \int_{D/2}^\infty 2\pi r h(r)dr \\
    &= 2\pi CD^B \int_{D/2}^\infty r^{1-B}dr \\
    &= \frac{\pi 2^{B-1}C}{B-2}D^{2-B} \\
    &\rightarrow C = \frac{D^{B-2}(B-2)}{\pi 2^{B-1}}V_\mathrm{c}
\end{align}

Note that $V_\mathrm{c}(D)$ can be calculated using the formulas in Section 1.2 above. For the most common case, $B=3$,

\begin{align}
    C &= \frac{D}{4\pi}V_\mathrm{c} \\
    h(r) &= \frac{D}{4\pi}V_\mathrm{c}\left(\frac{r}{D}\right)^{-3}
\end{align}

Ejecta may extend to great distances; however, once its thickness is smaller than the diurnal thermal skin depth, it is ineffective as an insulating (regolith) or conducting (rock) layer. This skin depth depends on the thermal diffusivity $\kappa$ and body's rotation period $P$,

\begin{align}
    z_\mathrm{s} = \sqrt{ \frac{\kappa P}{\pi} } 
\end{align}

Values of $z_\mathrm{s}$ for the moon range from $\sim 4$ cm (upper regolith) to $\sim 1$ m (rock). However, most of the ejecta volume is typically regolith, so we assume a constant value $z_\mathrm{s} \approx 10$ cm.

The "edge" of the ejecta blanket, $r_\mathrm{max}$ is the radial distance where $h(r) = z_\mathrm{s}$,

\begin{align}
    z_\mathrm{s} &= \frac{D}{4\pi}V_\mathrm{c}r_\mathrm{max}^{-3} \\
    r_\mathrm{max} &= \left( \frac{D}{4\pi z_\mathrm{s}}V_\mathrm{c} \right)^{1/3}
\end{align}

and

\begin{align}
    A_\mathrm{e}(D) = \left( \frac{\sqrt{\pi}D}{4 z_\mathrm{s}}V_\mathrm{c} \right)^{2/3}
\end{align}

#### 1.4 Crater production function

Impact fluxes have been estimated by various methods, as summarized by <i>Williams et al.</i> (2017). Figure 2 shows data from impact crater counting studies, in terms of the cumulative impactor flux, $N(>D)$.

```python
# Display figure from J-P. Williams
from IPython.display import Image
Image(filename='/Users/paha3326/main/research/collaborators/WilliamsJP/impact_flux/Figure_impactors_CSC.png')
```

**Figure 2:** Summary of impactor production function for the Moon, from *Williams et al. (2017)*.

---

From the figure above, we approximate a power-law relation for the cumulative distribution of impact crater size (gray boxes):

```python
import numpy as np
import matplotlib.pyplot as plt
import prettyPlots as pp
pp.setStyle()

# Model for crater production function
D = np.power(10,np.arange(0,4)) # Modeled crater diameters
CDF = [5e-3,4e-6,2e-9,1e-12] # Cumulative number [km-2.yr-1]
pfit = np.polyfit(np.log10(D),np.log10(CDF),1) # Power-law fit
Drange = np.power(10, np.linspace(-1,4,100))
Cfit = np.power(10, np.polyval(pfit, np.log10(Drange)))

# Plot
plt.figure(figsize=[6,8])
plt.loglog(D, CDF, 's', label='Williams $\mathrm{\it{et~al.}}$ (2017)')
plt.loglog(Drange, Cfit, '--', label='Power-law fit: {:.1f}'.format(pfit[0]))
plt.xlabel('Crater Diameter, $D$ (m)')
plt.ylabel('Cumulative Number $>D$ (km$^{-2}$ yr$^{-1}$)')
plt.legend(frameon=False, fontsize=16)
plt.savefig('/Users/paha3326/main/Research/moon/williams_impact-flux_power-law-fit.png', transparent=True, dpi=150, bbox_inches='tight')
plt.show()

print(pfit)
```

From the fit above, we find $\log_{10}N = -2.24 - 3.24\log_{10}D$. The PDF is then

\begin{align}
    n(D) = \frac{dN}{dD} = -0.0186 D^{-4.24}
\end{align}
